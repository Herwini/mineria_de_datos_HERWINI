```python
import pandas as pd

df = pd.read_csv("Mental_Health_Care_in_the_Last_4_Weeks.csv")
df.head()
```

```python
# observamos las variables categoricas y numericas
df.info()
```

# **Eliminacion de variables redundantes**
1. Variables como Time Period Label y Confidence Interval Time fueron eliminadas por una colinealidad alta.
2. La varaiable Supression Flag fue eliminada debido a que esta casi vacio en un 90%+ en el dataset.
3. La variable Quartile Range fue descartada para evitar el _data leakage_ en futuros modelos de clasificación o regresion, esta variable QR es un dato derivado de Value.

```python
# Eliminación de variables redundantes y ruidosas
df.drop(
    columns=[
        'Time Period Label',
        'Confidence Interval',
        'Quartile Range',
        'Suppression Flag'],
    inplace=True)

df.info()
```

# **Limpieza de registros nulos en la variable objetivo**

Garantizar la integridad de la variable dependiente Value, eliminamos las filas en donde haya valores faltantes.

```python
df.dropna(subset=['Value'], inplace=True)
df.info()
```

# Normalización de tipos de datos


```python
df['Time Period Start Date'] = pd.to_datetime(df['Time Period Start Date'])
df['Time Period End Date'] = pd.to_datetime(df['Time Period End Date'])
df.info()
```

# Guardamos el dataset con la limpieza hecha
El estudio a profundidad y problematicas de las demas variables queda pendiente para las practicas posteriores.

```python
df.to_csv("Mental_Health_Clean_Final.csv", index=False)
```
